```python
# Q.7) Create month name, your, day name,am/pm column

df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.dtypes
```

```python
df["Month_Name"] = df["Date"].dt.month_name()
df["Year"] = df["Date"].dt.year
df["Day_Name"] = df["Date"].dt.day_name()
df["AM/PM"] = df["Date"].dt.strftime("%I:%M:%S %p").str.split(" ").apply(lambda x: x[1])

df
```

```python
# Q.8) Show month wise income and Expense analysis Also use perfect chart

df.groupby(by = ["Month_Name","Income/Expense"])["INR"].sum().unstack().plot(kind = "bar",title= "Income Expence Analysis", figsize = (9,8), color = ["#ff6666", "#ffad33"])
```

```python
# Q.9) DAY NAME WISE expense analysis also show the chart

Expence = df[df["Income/Expense"] == "Expense"]
Expence.groupby(by = ["Day_Name","Income/Expense"])["INR"].sum().unstack().plot(kind = "bar",title= "Income Expence Analysis", figsize = (10,6), color = ["#80d4ff", "#ffccb3"])
```

```python
# Q.10) Show the top 5 categories with highest expense and top 5 categories with lowest expens ALSO SHOW chart

Expence = df[df["Income/Expense"] == "Expense"]
Expence.groupby(by = 'Category')["INR"].sum().nlargest(5).plot(kind = "bar", title= "Highest Expense", figsize = (8,6), color = ["#80d4ff", "#ffccb3"])
```

```python
Expence = df[df["Income/Expense"] == "Expense"]
Expence.groupby(by = 'Category')["INR"].sum().nsmallest(5).plot(kind = "pie",  autopct='%1.2f%%',title= "Expenses")
```
