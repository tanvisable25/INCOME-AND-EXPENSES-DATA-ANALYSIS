```python
data["Month_Name"] = data["Date"].dt.month_name()
data["Year"] = data["Date"].dt.year
data["Day_Name"] = data["Date"].dt.day_name()
data["AM/PM"] = data["Date"].dt.strftime("%I:%M:%S %p").str.split(" ").apply(lambda x: x[1])
```

```python
# Q.8) Show month wise income and Expense analysis Also use perfect chart

data.groupby(by = ["Month_Name","Income/Expense"])["INR"].sum().unstack().plot(kind = "bar",title= "Income Expence Analysis", figsize = (7,8), color = ["#ff6666", "#ffad33"])
```

```python
# Q.9) DAY NAME WISE expense analysis also show the chart

Expence = data[data["Income/Expense"] == "Expense"]
Expence.groupby(by = ["Day_Name","Income/Expense"])["INR"].sum().unstack().plot(kind = "bar",title= "Income Expence Analysis", figsize = (10,10), color = ["#80d4ff", "#ffccb3"])
```
